```python
import numpy as np
```

```python
array_2x2 = np.array([[1, 2],
                      [3, 4]])
print(array_2x2)
```

```python
rand_3x4 = np.random.rand(3, 4)
print(rand_3x4)
```

```python
ones_array = np.ones((2, 3))
zeros_array = np.zeros((2, 3))
print(ones_array)
print(zeros_array)
```

```python
rand_6x3 = np.random.rand(6, 3)
x= rand_6x3.T
print(rand_6x3)
print(x)
```

```python
n, p, m = 3, 2, 4
A = np.random.rand(n, p)
B = np.random.rand(p, m)
C= np.dot(A, B)
print(A)
print(B)
print(C)
```
